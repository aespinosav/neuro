# Spike Train Metrics

In this notebook we will show different ways in which one can construct a metric for spike--train space. The idea is that Spike Trains are objects that can be thought of as living in a metric space (not necessarily a normed one though).

We will introduce two types of approaches that can be shown to be related, and in some cases equivalent, to each other. The first is the kernel-based approach and the second is the edit-length approach.

We will mostly be following Houghton & Victor (2010).

## Kernel-based Metric

The main idea for the kernel based approaches is to think of each spike in the spike train as an instantaneous impulse (a Dirac $\delta$ function) and convolve it with a kernel function. The main procedure is similar to what is done for kernel density estimation.

By treating the spikes as delta functions, the expression for a whole spike train $\mathbf{u}$ (with $m$ spikes) becomes:
$$
g(t; \mathbf{u}) = \sum_{i=1}^{m} \delta(t - u_i),
$$ where $u_i$ is the time at which spike $i$ of the train occurs.
We can choose a kernel function $h_k(t)$, whose properties we will go into more detail later. Doing the convolution $h \star g$ we obtain:

\begin{align}
f(t) &= (g \star h)(t) \\
     &= \sum_{i=1}^{m} \int_0^T \delta(t - u_i) h_k(t) dt \\
     &= \sum_{i=1}^{m} \int_{u_i - \epsilon}^{u_i + \epsilon} \delta(t - u_i) h_k(t) dt\\
     &= \sum_{i=1}^{m} h_k(t - u_i)
\end{align}

The kernel function can be the boxcar kernel, the gaussian kernel or as we will be using an exponential kernel. The justification of using an exponential kernel lies in the physiological aspect of spike trains. So the kernel we will be using will be of the form,
$$
h_{\tau}(t) =
\left\{
\begin{matrix}
0 & t < 0\\
\dfrac{1}{\tau} e^{-t/\tau} & t\ge 0.
\end{matrix}
\right.
$$

Kernelising the spike train in this way is useful as it will provide us a way to calculate distances between different spike trains, i.e. how different they are. This is of interest since we will be analysing the spike trains of a neuron that result on exposure to the same stimulus.

Since we have now converted the spike train into a continuous, or at least piecewise-continuous function, the distance betwen two spikes, say $\mathbf{u}$ and $\mathbf{v}$, is given by the $L^{2}$ metric,
$$
d(\mathbf{u}, \mathbf{v}) = \sqrt{ \int dt \left[ f(t;\mathbf{u}) - f(t;\mathbf{v}) \right]^2}.
$$
