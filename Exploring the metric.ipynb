# Exploring the behaviour of the metric

In this notebook we explore the behaviour of the kernel based metric for spike trains. We will do so by defining a refference spike train composed of 2 spikes and measuring its distance to a test spike that has the first spike fixed to correspond with the first spike of the reference train but whose second spike is moved according to some parameter.

```python
s_ref = SpikeTrain(np.array([0.003, 0.006]), start_time=0, end_time=0.1)
```

```python
def test_spike(delta):
    spikes = np.array([0.003, 0.006 + delta*0.003])
    st = SpikeTrain(spikes, start_time=0, end_time=0.1)
    return st
```

```python
D = []
for d in np.arange(-2, 10, 0.01):
    dist = distance(s_ref, test_spike(d))
    D.append(dist)
```

```python
plt.plot(np.arange(-2,10,0.01), D)
```

```python
def test_spike2(delta):
    d = 0.0045
    spikes = np.array([0.003, d + 0.0015*delta ,0.006])
    st = SpikeTrain(spikes, start_time=0, end_time=0.1)
    return st
```

```python
D = []
for d in np.arange(-1, 1, 0.01):
    dist = distance(s_ref, test_spike2(d))
    D.append(dist)
```

```python
plt.plot(np.arange(-1, 1, 0.01), D)
```

```python
kern_sref = make_exp_kernel(s_ref, 0.003)
```

```python
plt.plot(np.arange(0,0.012, 0.0001), kern_sref(np.arange(0,0.012, 0.0001)))
```

```python
kern_s2 = make_exp_kernel(test_spike2(1), 0.003)
plt.plot(np.arange(0,0.012, 0.0001), kern_s2(np.arange(0,0.012, 0.0001)))
```

```python
plt.plot(np.arange(0,0.012, 0.0001), np.abs(kern_s2(np.arange(0,0.012, 0.0001)) - kern_sref(np.arange(0,0.012, 0.0001))))
         
```
